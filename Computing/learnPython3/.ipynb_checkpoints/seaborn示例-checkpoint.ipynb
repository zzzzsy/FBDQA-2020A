```python
import seaborn as sns
names = sns.get_dataset_names()
names
```

加载数据集

```python
tips = sns.load_dataset("tips")
tips
```

## 分布

### 直方图

```python
sns.displot(tips, x="total_bill" )
```

```python
sns.histplot(tips, x="total_bill")
```

```python
sns.displot(tips, x="total_bill",binwidth=3)  #bindwidth控制宽度
```

```python
sns.displot(tips, x="total_bill", bins=20)   ##bins表示分成多少个区间
```

所有绘图函数均接受 hue参数作为分类依据，hue根据颜色区分；

```python
sns.displot(tips, x="total_bill", hue="sex")
```

```python
sns.displot(tips, x="total_bill", hue="sex", multiple="stack")
```

图像级别的绘图函数接受 "col","row"参数作为划分子图的依据;

```python
sns.displot(tips, x="total_bill", hue="sex",col='sex')
```

```python
sns.displot(tips, x="total_bill",stat="density")  ##stat表示统计特征
```

### 概率密度图

使用高斯核函数根据已有数据估计概率分布

```python
sns.displot(tips, x="total_bill", kind="kde",hue='sex')
```

```python
sns.kdeplot(data=tips, x="total_bill",hue='sex')
```

```python
sns.displot(tips, x="total_bill", kind="kde", bw_adjust=.25)  #平滑参数
```

```python
sns.displot(tips, x="total_bill", kind="kde", bw_adjust=2)
```

```python
#也可以直方图和概率密度图同时显示
sns.displot(tips, x="total_bill", kde=True)
```

ecdf 累计概率分布，即小于当前所有值的概率分布，因此为单增曲线

```python
sns.displot(tips, x="total_bill", hue="sex", kind="ecdf")
```

```python
sns.ecdfplot(tips, x="total_bill", hue="sex")
```

```python
sns.displot(tips, x="total_bill", y="tip",hue='sex')
```

```python
sns.displot(tips, x="total_bill", y="tip", kind="kde")
```

## replot 关系图

## 散点图

```python
sns.relplot(x="total_bill", y="tip", data=tips)  
```

```python
sns.scatterplot(data=tips,x="total_bill", y="tip")
```

style参数根据类别使用不同的形状，和hue参数可以相同也可以不同；尽量使用同一变量，否则可能难以区分

```python
sns.relplot( data=tips,x="total_bill", y="tip", hue="smoker", style="smoker")
```

```python
sns.relplot( data=tips,x="total_bill", y="tip", hue="smoker", style="sex")
```

```python
sns.scatterplot(data=tips,x="total_bill", y="tip",hue="smoker", style="smoker")
```

```python
sns.relplot(x="total_bill", y="tip", hue="size", palette="ch:r=-.5,l=.75", data=tips)  #可以设置渐变色的范围
```

```python
sns.relplot(x="total_bill", y="tip", size="size", data=tips)   #size参数使用不同大小的形状区分类别
```

```python
fmri = sns.load_dataset("fmri")
fmri
```

### 折线图

折线图一般用来表示变量随时间的变化关系，会将同一时间的数据按照置信度画出重叠区域，可通过ci参数设置

```python
sns.relplot(data=fmri,x="timepoint", y="signal",hue="event", style="event",kind="line",ci=None)
```

```python
sns.relplot(data=fmri,x="timepoint", y="signal",hue="event", style="event",kind="line")
```

```python
sns.lineplot(data=fmri,x="timepoint",hue="event", style="event",y="signal")
```

```python
sns.relplot(data=tips,x="total_bill", y="tip", hue="sex",col='smoker')
```

### 联合分布&边缘分布

```python
sns.jointplot(data=tips, x="total_bill", y="tip",hue='sex')
```

```python
sns.pairplot(data=tips,hue='sex')
```

## Catgory Plot

```python
sns.catplot(data=tips,x="day",y="tip")
```

```python
sns.catplot(data=tips,x="day",jitter=False,y="tip")
```

```python
sns.stripplot(data=tips,x="day",jitter=False,y="tip")
```

```python
sns.catplot(data=tips,x="day", y="tip", kind="swarm")
```

```python
sns.swarmplot(data=tips,x="day", y="tip")
```

```python
sns.catplot(x="smoker", y="tip", order=["No", "Yes"], data=tips)
```

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips,hue="smoker")
```

```python
sns.catplot(x="day", y="total_bill", kind="boxen", data=tips)
```

```python
sns.catplot(x="day", y="tip", kind="violin", data=tips)
```
