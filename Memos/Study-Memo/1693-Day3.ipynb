```python
#2. 猫（颜色color，动作jump）类的定义和实现：
class cat:
    def __init__(self,color,location):
        self.color=color
        self.location=location
    def jump(self,distance):
        self.location+=distance
```

```python
maomao=cat('black',0)
maomao.jump(2)
maomao.color
```

```python
maomao.location
```

```python
#4. 余弦函数绘制：
import numpy as np
xx=np.arange(0,2,0.01)
yy=np.cos(2*np.pi*xx)
plt.plot(xx,yy,color='b')
```

```python
#余弦函数的线图：
fig,ax=plt.subplots()
ax.plot(xx,yy)
ax.set(xlabel='time(s)',ylabel='voltage(mV)',title='Voltage/Time')
ax.grid()
#fig.savefig("test.png")
plt.show()
```

```python
#5. iris数据集联合分布和边缘分布（联合分布使用回归分析）：
import seaborn as sns
data_iris=sns.load_dataset('iris')
data_iris.info()
```

```python
sns.pairplot(data=data_iris,hue='species',kind='reg')
```

```python
sns.jointplot(x="sepal_length",y="sepal_width",data=data_iris,kind='reg')
```

```python
sns.pairplot(data=data_iris,hue='species',kind='scatter')
```

```python
sns.jointplot(x="sepal_length",y="sepal_width",data=data_iris,kind='hex')
```
